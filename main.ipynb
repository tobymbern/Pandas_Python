```python
import pandas as pd
import numpy as np
students = pd.read_csv('students_complete.csv')
schools = pd.read_csv('schools_complete.csv')
```

```python
schools.head()
```

## District Summary

```python
columns = ['Total Schools','Total Students','Total Budget','Average Math Score',
           'Average Reading Score','% Passing Math','% Passing Reading','Overall Passing Rate']
index = [0]
district_summary = pd.DataFrame(index = index, columns=columns)
num_students = students['Student ID'].count()
num_schools = schools['School ID'].count()
passing_math = students[students.math_score > 69]['math_score'].count()
passing_reading = students[students.reading_score > 69]['reading_score'].count()

# I'm assuming that below 70% is failing.
district_summary['Total Schools'] = num_schools
district_summary['Total Students'] = num_students
district_summary['Total Budget'] = schools['budget'].sum()
# I'm assuming that a passing grade is 70%
district_summary['Average Math Score'] = students['math_score'].mean()
district_summary['Average Reading Score'] = students['reading_score'].mean()
district_summary['% Passing Math'] = passing_math/num_students
district_summary['% Passing Reading'] = passing_reading/num_students
district_summary['Overall Passing Rate'] = (passing_math + passing_reading)/(num_students*2)
district_summary.head()
```

## School Summary

```python
columns2 = ['School Name','School Type','Total Students','Total School Budget',
            'Per Student Budget','Average Math Score','Average Reading Score',
            '% Passing Math','% Passing Reading','Overall Passing Rate']
index2 = np.arange(0,num_schools)
school_summary = pd.DataFrame(index = index2, columns = columns2)
school_summary['School Name'] = schools.name
school_summary['School Type'] = schools.type
school_summary['Total Students'] = schools['size']
school_summary['Total School Budget'] = schools.budget
school_summary['Per Student Budget'] = schools.budget/schools['size']
school_summary['Average Math Score'] = [students[students.school == row]['math_score'].mean()
                                        for row in schools['name']]

school_summary['Average Reading Score'] = [students[students.school == row]['reading_score'].mean()
                                        for row in schools['name']]

school_summary['% Passing Math'] = [students[(students.school == row) & (students.math_score > 69)]['math_score'].count()/ 
                                    students[students.school == row]['math_score'].count() 
                                    for row in schools['name']]

school_summary['% Passing Reading'] = [students[(students.school == row) & (students.reading_score > 69)]['reading_score'].count()/ 
                                       students[students.school == row]['reading_score'].count() 
                                       for row in schools['name']]

school_summary['Overall Passing Rate'] = (school_summary['% Passing Math'] + school_summary['% Passing Reading'])/2
                                        
school_summary.head(15)

```

## Top Peforming Schools (By Passing Rate)

```python
list_top_schools = school_summary.nlargest(5, 'Overall Passing Rate')
list_top_schools.head()
```

## Bottom Performing Schools (By Passing Rate)

```python
list_bottom_schools = school_summary.nsmallest(5, 'Overall Passing Rate')
list_bottom_schools.head()
```
